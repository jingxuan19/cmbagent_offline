```python
from pathlib import Path
from cmbagent.workflows import one_shot
import os
# BASE_URL = os.getenv("LOCAL_LLM_BASE_URL")
BASE_URL = "http://localhost:8010/v1"
# API_KEY = os.getenv("LOCAL_LLM_API_KEY")
API_KEY = "EMPTY"

from dotenv import load_dotenv
dotenv_path = Path("/scratch/scratch-lxu/venv/cmbagent_env/.env")
load_dotenv(dotenv_path)


local_llm_config = {
    "config_list": [
        {
            "model": "gpt-oss-120b", # Must match model id from endpoint
            "api_key": API_KEY,
            "api_type": "openai",
            "base_url": BASE_URL,
            "price": [0.0, 0.0], # Passed explicitly to ignore warning
        },
    ],
    "cache_seed": None
}

local_engineer_config = local_llm_config["config_list"][0]

tmp_path = Path.cwd()
```

```python
task = """
There are 35 green balls, 25 blue balls, and 40 red balls in a box.
I draw the balls one by one without replacement until there are two colors left.
What is the expected number of balls remaining in the box?

Suppose there are G green balls, R red balls, and B blue balls,
plot a graph of this expected number as a function of the G, B, R respectively.
Show them on a single plot.
"""

results = one_shot(
    task,
    max_rounds=5,
    agent="engineer",
    engineer_model=local_engineer_config,
    default_llm_model=local_engineer_config,       # For controller, executor, etc.
    default_formatter_model=local_engineer_config, # For response formatters
    work_dir=str(tmp_path / "one_shot"),
)
```
